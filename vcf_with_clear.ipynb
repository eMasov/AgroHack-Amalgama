```python
import numpy as np
X = df_clear.values
X_trunc = X.copy()
y = list(data_phen["2019"])
inds = []
for i in range(len(y)):
  if np.isnan(y[i]):
    inds.append(i)
X_trunc = np.delete(X_trunc,inds,axis = 0)
y = np.delete(y,inds,axis = 0)
```

```python
X_trunc.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_trunc,y,test_size = 0.15,random_state = 42)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import r2_score
ctb = CatBoostRegressor(iterations = 500,cat_features=list(range(0,38940)),task_type="GPU")

ctb.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print(r2_score(y_test,ctb.predict(X_test)))
print(mean_absolute_percentage_error(y_test,ctb.predict(X_test)))
```

```python
imps = ctb.get_feature_importance()
imps[imps > 0]
```
